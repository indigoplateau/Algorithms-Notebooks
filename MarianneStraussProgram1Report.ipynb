# Part 2 - Pseudo-Code

#### Bubble Sort
This is an ineffiecient and naive sorting algorithm with expected running time of $O(n^2)$.
This is done by the algorithm by repeatedly swapping adjacent elenents that are out of order.

```

BUBBLESORT(A):
1  For i = 1 to A.length - 1
2     for j = A.length downto i + 1
3         if A[j] < A[j - 1]
4             exchange A[j] with A[j - 1]

```
Note pseudocode referenced from textbook. See citations.

#### Merge Sort

This is a sorting algorithm that implements a divide and conquer strategy with an epected running time of $O(nlogn)$.
This achieved by dividing the n-element array to be sorted into two subarrays. Both subarrays are then sorted recursively sorted using merge sort. Then the two subarrays are merged to form a sorted array.

The following pseudocode is based on the textbook's pseudocode and definition of mergesort(see citations). 

Below is the subroutine function that merges array, A, with the indices p,q,r where $p <= q < r$. This function assumes subarrays $A[p ..q]$ and $A[q+1..r]$ are already in sorted order. It merges the sorted subarrays and replaces the current subarray $A[p ..r]$ into one sorted subarray.

```
MERGE(A, p, q, r)
1 n1 = q - p + 1
2 n2 = r - q
3 let L[1..n1 + 1] and R[1..n2 + 1] be new arrays
4 for i = 1 to n1
5    L[i] = A[p + i - 1]
6 for j = 1 to n2
7    R[j] = A[q + j]
8 L[n1 + 1] = sentinel(infinity)
9 R[n2 + 1] = sentinel(ininity)
10 i = 1
11 j = 1
12 for k = p to r
13     if L[i] <= R[j]
14         A[k] = L[i]
15         i = i + 1
16     else A[k] = R[j]
17         j = j + 1
```
The function Merge-Sort sorts the elements in the subarray $A[p ..r]$. If $p >= r$ then the subarray has at most one element and is therefore sorted.  Otherwise the function partitions $A[p ..r]$ into two subarrays, $A[p ..q]$ and $A[q + 1 ..r]$ and each subarray contains $floor(n/2)$ elements. The first recursive call of Merge-Sort in MergeSort to sort array A. The algorithm works by mergeing pairs of sorted sequences of length 1, then 2, then 3, and so on until the merged sorted sequences of length n/2 is of length n.

```
MERGE-SORT(A, p, r)
1 if p < r
2    q = floor((p + r)/2)
3    MERGE-SORT(A, p, q)
4    MERGE-SORT(A, q + 1, r)
5    MERGE(A, p, q, r)
```
#### Counting Sort

This sorting algorithm is an integer sorting algorithm that assumes that the input array or sequence contains only real numbers or an integer in range $0 to k$. It then determines for each array element x, the number of array elements less the the key value, element x. It then uses this number to place x directly into its final position in the output array. The number of occurences of an element is stored/hashed in a temporary counting array.

The above definition and the pseudocode is based on the reference of the textbook and the article outlining Counting Sort on the website geeksforgeeks (see citations).

```
COUNTING-SORT(A, B, k)
1 let C[0 ..kj] be a new array
2 for i = 0 to k
3     C[i] = 0
4 for j = 1 to A.length
5     C[A[j]] = C[A[j]] + 1
6 for = 1 to k
7     C[i] = C[i] + C[i - 1]
8 for j = A.length downto 1
8     B[C[A[j]]] = A[j]
9     C[A[j]] = C[A[j]] - 1
```
A is the input array, B is the output array and k is range of the input or elements in array A.

# Part 3 - Static Analysis

## Bubble Sort Loop Invariant

#### Loop invariant:
At the start of each iteration of the for loop of lines 1 - 4 in part 2, the subarray $A[1 .. i - 1]$ consists of $i - 1$ smallest elements in $A[i ..n]$ which are in sorted order. The subarray $A[i ..n]$ consists of the rest of elements in some order.

##### Initialization:
Loop invariant holds before the first loop iteration because $A[1 ..i-1]$ is empty and therefore trivially sorted and the smallest.

##### Maintenance:
Assume that prior to an iteration of the for loop in lines 1-4, the subarray $A[1 ..i - 1]$ consists of $i -1$ smallest elements in $A[i ..n]$ are in sorted order. We need to show that it is also true pror to the next iteration.

If the inner loop in lines 2-4 completely iterates, if $A[j] < A[j-1]$ they will be swapped or left as is and $A[i]$ will be the smallest element in subarray $A[i ..n]$. Therefore before the next iteration of the main outer loop in lines 1-4 the subarray $A[1 ..i - 1]$ consists of smaller elements than in subarray $A[i + 1 ..n]$  in sorted order.

##### Termination
When i in the for loop equals $A.length$ the outer loop terminates. If $i = A.Length = n$ and i iterates by 1 then at the time of termination then $i = n + 1$. When you substitute $i = n + 1$ into the loop invariant the subarray $A[1 .. i - 1]$, the subarray is then $A[1 ..n]$ which means the entire array is sorted and the sorting algorithm is correct.



## Merge Sort Loop Invariant

### *NOTE the below loop invariant is based on textbook explaination. See citations.

#### Loop invariant: 
At the start of each iteration of the for loop in lines 12-17 in part 2, the subarray $A[p ..k-1]$ consists  of $k - p$ smallest elements of the arrays $L[1 ..n1 + 1]$ and $R[1 ..n2 + 1]$ which are in sorted order.

##### Initialization:
Loop invariant holds before the first loop iteration because $p = k$ and $A[p ..k - 1]$ is empty and therefore trivially sorted.

##### Maintenance:
Assume that prior to an iteration of the for loop in lines 12-17, the subarray $A[p ..k-1]$ consists  of $k - p$ smallest elements of the arrays $L[1 ..n1 + 1]$ and $R[1 ..n2 + 1]$ which are in sorted order.

If $L[i] <= R[j]$, and $L[i]$ is assigned to $A[k]$, the subarray $A[p ..k]$ will consist of $k - p + 1$ smallest elements. If $L[i] > R[j]$, $R[j]$ is assigned to $A[k]$; both actions will maintain the loop invariant in either case.

##### Termination
When k in the for loop equals $r + 1$ the loop terminates.  Because of the loop invariant $A[p ..k-1]$ consists  of $k - p$ smallest elements of the arrays $L[1 ..n1 + 1]$ and $R[1 ..n2 + 1]$ in sorted order. When substituting $k = r + 1$ into the the loop invariant, $k - p = r - p + 1$. Therefore arrays L and R have all of the contents copied back into A except the sentinel values (infinity).




## Counting Sort

#### Loop invariant: 
At the start of each iteration of the for loop in lines 10-12 in part 2, array $B[1 ..n]$ contains the sorted ouput and array $C[0 ..k]$ contains the remaining positions of each integer element in array $A[1 ..n]$.

##### Initialization:
Loop invariant holds before the first loop iteration because $B[1 ..n]$  is empty and therfore trivially contains the sorted ouput. array $C[0 ..k]$ contains the unchanged positions of each integer element in array $A[1 ..n]$. 

##### Maintenance:
Assume that prior to an iteration of the for loop in lines 10-12, that array $B[1 ..n]$ contains the sorted ouput and array $C[0 ..k]$ contains the remaining count of each integer element in array $A[1 ..n]$.

After element $A[j]$ is placed into its correct sorted postion into $B[C[A[j]]]$, $C[A[j]]$ is then decremented. The Loop invariant then still holds true before next iteration.

##### Termination
When $j = 1$ all elements of array $A[1 ..n]$ have been placed in the correct sorted order in output array $B[1 ..n]$ and therefore the sorting algorithm is correct.

# Part 4 - Implementation

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

import operator
import math
import time
````

Below is the function used to count number of comparisons and a global comparison counter.

````python
n_comparisons = 0


def compare(f, x, y):
    global n_comparisons
    n_comparisons += 1
    return f(x,y)
````

Below is the function used to count number of assignments and a global assignment counter.

````python
n_assignments = 0

def assign(a, i, x):
    global n_assignments
    n_assignments += 1
    a[i] = x
````

#### *NOTE the above functions are based on code from Assignment 0 Sample report and refactored using operator module, works cited.

### Bubble Sort Implementation

````python
def bubble_sort(a):
    global n_assignments
    n_assignments = 0
    n = len(a)
    
    for i in range(n):
        for j in range(n, i, -1):
            if compare(operator.lt, a[j-1], a[j-2]):
                
                temp = a[j-2]
                assign(a, j-2, a[j-1])
                assign(a, j-1, temp)
        
````

#### Testing of Bubble Sort

````python
test= [4,3,5,1,7]

bubble_sort(test)
print(test)

string = "Number of comparisons: "
print(string, n_comparisons)
string = "Number of assignments: "
print(string, n_assignments)
````

### Merge Sort Implementation


````python
def merge(a, p, q, r):
    
    n1 = q - p + 1
    n2 = r - q
    
    L = [0] * (n1)
    R = [0] * (n2)
    
    for i in range(0, n1):
        assign(L, i, a[p + i])
        
    for j in range(0, n2):
        assign(R, j, a[q + j + 1])
    
    L.append(math.inf)
    R.append(math.inf)
    
    i = 0
    j = 0
    
    for k in range(p, r + 1):

        if (compare(operator.le,L[i], R[j])):
            assign(a, k, L[i])
            i += 1
        else:
            assign(a, k, R[j])
            j += 1
    
````

````python
global n_comparisons
global n_assignments
n_comparisons = 0
n_assignments = 0

def merge_sort(a, p, r):
    
    if (p < r):
        q = (p + r)//2
        merge_sort(a, p, q)
        merge_sort(a, q + 1, r)
        merge(a, p, q, r)
        
````

````python
test= [5,2,4,7,1,3,2,6,8]
n = len(test)

merge_sort(test, 0, n - 1)
print("test:")
print(test)

string = "Number of comparisons: "
print(string, n_comparisons)
string = "Number of assignments: "
print(string, n_assignments)
````

### Counting Sort Implementation

````python
def counting_sort(a, b, k):
    global n_comparisons
    global n_assignments
    n_comparisons = 0
    n_assignments = 0
    
    
    n = len(a)
    C = [0] * (k + 1)
    
    for i in range(0, n):
        C[a[i]] += 1
    
    for i in range(1, k+1):
        C[i] = C[i] + C[i - 1]
    
    for i in range(n - 1, -1, -1):
      
        assign(b, C[a[i]]- 1, a[i] )
        C[a[i]] -= 1
        
````

````python
test= [5,2,4,7,1,3,2,6,8]
n = len(test)
output = [0] * n
k = np.max(test)
counting_sort(test, output, k)

print("test:")
print(output)

string = "Number of comparisons: "
print(string, n_comparisons)
string = "Number of assignments: "
print(string, n_assignments)
````

# Part 5 - Analysis

````python
def readFile(filename, n = 0):
    with open(filename, 'r') as reader:
        
        description = reader.readline()
        n = int(reader.readline())
        a = [0] * n
        
        i = reader.readline()
        j = 0;
        while i != '':
            a[j] = int(i)
            j += 1
            
            i = reader.readline()
        
        return description, a
````

### One Million Randoms

Here we analize each algorithm using a dataset of one million random integer values from 1 to 100.

The cell below I am prepping an array to be copied for each algorithm to sort by reading in the values, a dataframe, and a variables for storing time.

````python
startTime = 0
endTime = 0
finalTime = 0

df_index = 0
comparison_dataframe = pd.DataFrame(columns = ['Algo Name', 'Comparisons', 'Assignments', 'Run Time (ns)', 'Number of Elements'])

## Reading in File
description, a = readFile('one-million-randoms.txt')
#description, a = readFile('shuffled.txt')
````

### Bubble Sort with  with 1 to 100,000 Randoms

````python
n_comparisons = 0
n_assignments = 0

n = 1

while (n <= 10000):
    
    #### Running Bubble Sort for n elements ####
    
    a1 = a[:n]
    
    startTime = time.time_ns()
    bubble_sort(a1)
    #print(a1)
    
    endTime = time.time_ns()
    finalTime = (endTime - startTime)

    ## appending data from bubble sort for n = 1 to dataframe

    comparison_dataframe.loc[df_index] = ['Bubble Sort', n_comparisons, n_assignments, finalTime, n]
    
    n *=  10
    df_index += 1

comparison_dataframe
````

## NOTE
The above algorithm (Merge Sort) took greater than 1 hour to run for one million elements and only sorted a fraction of the dataset . Therefore I have no included this data in the dataframe used to analyze and compare each sorting algorithm. This is because the worse case run time complexity is $O(n^2)$. The algorithm can be further optimized like mentioned above by adding tracking if a swap occurs in the nested for loop and in the condition that $A[j] < A[j-1]$ is true. The flag is switched to $True$ if $A[j] < A[j-1]$ and a swap occurs but if the nested loop runs and is never swapped the algorithm is stopped. This improves only the running time by constant values and thefore it is still at worst $O(n^2)$.

### Merge Sort with 1 to 1,000,000 Randoms

````python
n_comparisons = 0
n_assignments = 0

n = 1

while (n <= 1000000):
    #### Running Merge Sort  for n elements####

    a2 = a[:n]
    num = len(a2)

    startTime = time.time_ns()

    merge_sort(a2, 0, num - 1)
    #print(a2)

    endTime = time.time_ns()
    finalTime = (endTime - startTime)

    ## appending data from merge sort of n elements to dataframe

    comparison_dataframe.loc[df_index] = ['Merge Sort', n_comparisons, n_assignments, finalTime, n]
    
    n *=  10
    df_index += 1
    
comparison_dataframe
````
